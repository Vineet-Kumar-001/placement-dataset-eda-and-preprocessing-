```python
df['part_time_job_num'] = df['part_time_job'].map({'Yes':1, 'No':0})
```

```python
df1 =  df.groupby(['branch','gender'])['part_time_job_num'].sum()
```

```python
df1
```

```python
df_plot = df1.reset_index()
```

```python
sns.barplot(data=df_plot,x='branch',y='part_time_job_num',hue='gender',palette='rainbow')
plt.title('Part-time Jobs by Branch and Gender')
plt.ylabel("part time job done by student ")
plt.xlabel("Branch")
plt.show()
```

```python
stress_levels_by_branch = df.groupby(['branch','gender'])['stress_level'].sum()
stress_levels_by_branch
```
